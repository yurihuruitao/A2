```python
import os
import re

input_file = '儒林外史.txt'   # 源文件名
out_dir = 'chapters'           # 输出文件夹

os.makedirs(out_dir, exist_ok=True)

with open(input_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

chapters = []
current = None

for line in lines:
    # 如果行首（忽略前导空白）是*，认为是新章节的标题
    if re.match(r'^\s*\*', line):
        if current is not None:
            chapters.append(current)
        current = [line]
    else:
        if current is None:
            # 如果文件开头没有遇到 * 就把前导内容也作为第一个章节的一部分
            current = [line]
        else:
            current.append(line)

# 把最后一章加入列表
if current is not None:
    chapters.append(current)

# 写出每一章到单独的文件
for i, ch in enumerate(chapters, start=1):
    out_path = os.path.join(out_dir, f'Chapter{i}.txt')
    with open(out_path, 'w', encoding='utf-8') as out_f:
        out_f.writelines(ch)

print(f'已保存 {len(chapters)} 章到文件夹：{out_dir}')
```
